<h2>Reducci√≥n de Dimensionalidad</h2>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Asumiendo que la columna 'cid' es un identificador y no debe ser incluida en el PCA
features = df.drop(columns=['cid'])
# Estandarizar los datos
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
# Aplicar PCA
pca = PCA()
pca_features = pca.fit_transform(scaled_features)
# Tomar las primeras dos componentes principales para el gr√°fico de dispersi√≥n
pc1 = pca_features[:, 0]
pc2 = pca_features[:, 1]
# Gr√°fico de dispersi√≥n
plt.figure(figsize=(10, 7))
plt.scatter(pc1, pc2, alpha=0.6, edgecolors='w', linewidth=0.5)
plt.title('Gr√°fico de dispersi√≥n de las dos primeras componentes principales')
plt.xlabel('Primera componente principal (PC1)')
plt.ylabel('Segunda componente principal (PC2)')
plt.grid(True)
plt.show()
```

 Aunque hay cierta dispersi√≥n, no se observa una clara separaci√≥n entre posibles grupos o clusters en este espacio bidimensional. Sin embargo, este es solo un vistazo inicial y se pueden realizar an√°lisis m√°s profundos para identificar patrones o grupos en los datos. Como el An√°lisis de Varianza

```python
#An√°lisis de Varianza
# Calcular la varianza explicada y la varianza acumulada explicada
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance)

# Gr√°fico de la varianza explicada y acumulada
fig, ax1 = plt.subplots(figsize=(14, 7))

# Bar plot para cada varianza individual
ax1.bar(range(len(explained_variance)), explained_variance, alpha=0.6, align='center', label='Varianza explicada individual', color='g')
ax1.set_xlabel('Componentes principales')
ax1.set_ylabel('Varianza explicada', color='g')
ax1.tick_params('y', colors='g')

# Line plot para cada varianza cumulativa
ax2 = ax1.twinx()
ax2.plot(range(len(explained_variance)), cumulative_explained_variance, label='Varianza explicada acumulada', color='b', marker='o')
ax2.set_ylabel('Varianza explicada acumulada', color='b')
ax2.tick_params('y', colors='b')

plt.title('Varianza explicada por las componentes principales')
fig.tight_layout()
plt.show()
```

El gr√°fico muestra la varianza explicada por cada componente principal (barras verdes) y la varianza acumulada explicada (l√≠nea azul con marcadores).

Como se puede observar, las primeras componentes principales capturan la mayor parte de la varianza de los datos. La l√≠nea azul muestra c√≥mo la varianza explicada acumulada aumenta a medida que se consideran m√°s componentes. Por ejemplo, alrededor de las primeras 10 componentes, ya se ha capturado una gran parte de la varianza total de los datos.

Este tipo de an√°lisis es √∫til para decidir cu√°ntas componentes principales conservar al reducir la dimensionalidad. Por ejemplo,  el 95% de la varianza.

Capturar el 95% de la varianza implica seleccionar el n√∫mero m√≠nimo de componentes principales que suman hasta ese porcentaje de varianza explicada. Una vez que se haya determinado ese n√∫mero, se puede reducir la dimensionalidad de los datos a ese n√∫mero de componentes y luego aplicar K-Means u otros algoritmos de clustering.

Pasos:

*Determina el n√∫mero de componentes principales necesarias para capturar el 95% de la varianza.

*Reducir la dimensionalidad de tus datos a ese n√∫mero de componentes.

*Aplica K-Means para agrupar los datos en clusters.

```python
# Determinar el n√∫mero de componentes para capturar el 95% de la varianza
num_components = np.where(cumulative_explained_variance >= 0.95)[0][0] + 1  # +1 porque los √≠ndices comienzan en 0

# Reducir la dimensionalidad a ese n√∫mero de componentes
pca_95 = PCA(n_components=num_components)
reduced_data_95 = pca_95.fit_transform(scaled_features)

num_components
```

 Se 14 componentes principales para capturar al menos el 95% de la varianza en tus datos.

<h2>Clustering</h2>

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# Rango de clusters a evaluar
range_clusters = range(2, 21)

inertia_95 = []
silhouette_scores_95 = []
# Determinar la inercia y el coeficiente de silueta para cada n√∫mero de clusters
for n_clusters in range_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(reduced_data_95)
    labels = kmeans.labels_

    inertia_95.append(kmeans.inertia_)
    silhouette_scores_95.append(silhouette_score(reduced_data_95, labels))

# Graficar el m√©todo del codo y la silueta para los datos reducidos
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# M√©todo del codo
ax1.plot(range_clusters, inertia_95, marker='o', linestyle='--')
ax1.set_title('M√©todo del Codo (95% de varianza)')
ax1.set_xlabel('N√∫mero de Clusters')
ax1.set_ylabel('Inercia')
ax1.grid(True)

# Silueta
ax2.plot(range_clusters, silhouette_scores_95, marker='o', linestyle='--', color='r')
ax2.set_title('Score de Silueta (95% de varianza)')
ax2.set_xlabel('N√∫mero de Clusters')
ax2.set_ylabel('Silhouette Score')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

Aplicando 4 Clusters

```python
# Aplicar K-Means con 4 clusters a los datos reducidos
kmeans_4_clusters = KMeans(n_clusters=4, random_state=42)
clusters = kmeans_4_clusters.fit_predict(reduced_data_95)

# Gr√°fico de dispersi√≥n con los clusters identificados
plt.figure(figsize=(10, 7))
plt.scatter(reduced_data_95[:, 0], reduced_data_95[:, 1], c=clusters, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=0.5)
plt.title('Clusters identificados con K-Means (4 clusters)')
plt.xlabel('Primera componente principal (PC1)')
plt.ylabel('Segunda componente principal (PC2)')
plt.grid(True)
plt.colorbar()
plt.show()
```

<h2>Evaluando M√©tricas</h2>

```python
# Calcular inercia y coeficiente de silueta para los clusters generados
kmeans_inertia = kmeans_4_clusters.inertia_
silhouette_avg = silhouette_score(reduced_data_95, clusters)

kmeans_inertia, silhouette_avg
```

Inercia (o SSE, Sum of Squared Errors): Este valor representa la suma de las distancias al cuadrado de cada punto de datos a su centroide de cluster m√°s cercano. Es una medida de cu√°n dispersos est√°n los datos dentro de cada cluster. Un valor m√°s bajo de inercia es mejor, ya que indica que los puntos de datos est√°n m√°s cerca de los centroides de sus clusters. Sin embargo, la inercia tiende a disminuir a medida que aumentamos el n√∫mero de clusters, incluso si esos clusters no son significativos, por lo que es importante equilibrarla con otras m√©tricas y el conocimiento del dominio.

En este caso, la inercia es
4905.722930573602


Coeficiente de Silueta: Esta m√©trica mide qu√© tan similar es un objeto a su propio cluster (cohesi√≥n) en comparaci√≥n con otros clusters (separaci√≥n). Los valores del coeficiente de silueta oscilan entre -1 y 1, donde un valor alto indica que el objeto est√° bien emparejado con su propio cluster y mal emparejado con los clusters vecinos. Si la mayor√≠a de los objetos tienen un valor alto, entonces la configuraci√≥n de clusters es apropiada. Si muchos puntos tienen un valor bajo o negativo, entonces la configuraci√≥n de clusters puede tener demasiados o muy pocos clusters.

En este caso, el coeficiente de silueta es  0.3806446928415913. Este valor es relativamente bueno ya que est√° cerca de 0.5, lo que indica una estructura razonable de clusters. Sin embargo, a√∫n hay margen de mejora.
