<h2>Técnicas de Imputación para tratar Nulls</h2>

*   Elemento de lista



---

La variable **target** en nuestro conjunto de datos de cáncer es categórica y crucial para la predicción. Contiene 6 valores nulos, que representan aproximadamente el 1% del total de registros. Dado que esta cantidad es trivial y la imputación podría sesgar un futuro modelo de aprendizaje automático, optamos por eliminar estos registros con valores nulos.

Opciones comunes para la imputación:

-**Media o Mediana**: Para características numéricas, una opción común es imputar los valores nulos con la media o la mediana de la columna. Especialmente útil cuando los datos faltantes son aleatorios.

-**Moda**: Para características categóricas, la moda (valor más frecuente) suele utilizarse para imputar valores nulos.

Dado que todas las variables independientes con valores nulos son numéricas y este es un conjunto de datos médicos (donde la media o la mediana podrían tener un significado clínico), **se propone usar la mediana para la imputación**. La mediana es robusta a los valores atípicos, lo que podría ser útil en un contexto médico donde los valores extremos pueden ser posibles pero no necesariamente indicativos de la población general.

```python
df_inicial = df_inicial.drop(['Unnamed: 0'], axis=1)
df_inicial = df_inicial.dropna(subset=['target']) #Se elimina los registros de con valores nulos de la variable target
df_inicial = df_inicial[df_inicial['target'].isin([0, 1])] #Solo se queda con categorias 1 y 0 de la variable target
df = df_inicial.fillna(df_inicial.median()) ## Hallando la mediana para todas las variables y rellenando los valores nulos
print(f'La base de datos (Luego de las técnicas de imputación) tiene un tamaño de {len(df)} registros')
print('Aquí se puede observar los valores nulos para cada variable luego de la imputación')
df.isnull().sum() # Se verifica si existe un valor nulo en cada variable
```

```python
df['target'] = df['target'].map({0: 'No Cancer', 1: 'Cancer'})
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='target')
plt.title('Distribución de la variable Target')
plt.xlabel('Categorías')
plt.ylabel('Número de registros')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')
plt.show()
```

### Tratamiento de Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir el número de filas y columnas para la disposición de los gráficos
num_rows = 5
num_cols = 6

# Crear una figura y ejes para los subgráficos
fig, axs = plt.subplots(num_rows, num_cols, figsize=(18, 12))
axs = axs.ravel()

# Lista de características numéricas
numeric_features = [
    'mean radius', 'mean texture', 'mean perimeter', 'mean area',
    'mean smoothness', 'mean compactness', 'mean concavity',
    'mean concave points', 'mean symmetry', 'mean fractal dimension',
    'radius error', 'texture error', 'perimeter error', 'area error',
    'smoothness error', 'compactness error', 'concavity error',
    'concave points error', 'symmetry error', 'fractal dimension error',
    'worst radius', 'worst texture', 'worst perimeter', 'worst area',
    'worst smoothness', 'worst compactness', 'worst concavity',
    'worst concave points', 'worst symmetry', 'worst fractal dimension'
]

# Iterar a través de las características y crear los gráficos de boxplot
for i, feature in enumerate(numeric_features):
    row = i // num_cols
    col = i % num_cols

    sns.boxplot(x='target', y=feature, data=df, ax=axs[i], palette='pastel')
    axs[i].set_xlabel('Target')
    axs[i].set_ylabel(feature)
    axs[i].set_title(f'Boxplot of {feature}')
    axs[i].grid(True)

    # Eliminar ejes si no se completan todas las celdas
    if i >= len(numeric_features) - num_cols:
        axs[i].axis('off')

# Ajustar el diseño de la figura
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

```python
# Crear histogramas de las variables numéricas
num_rows = 5
num_cols = 6

fig, axs = plt.subplots(num_rows, num_cols, figsize=(18, 12))
axs = axs.ravel()

for i, var in enumerate(numeric_features):  # Cambiar a numeric_features
    row = i // num_cols
    col = i % num_cols

    axs[i].hist(df[var], bins=20, color='blue', alpha=0.7)
    axs[i].set_title(('Histogram of ' + var))
    axs[i].set_ylabel('Frequency')
    axs[i].set_xlabel(var)
    axs[i].grid()



plt.tight_layout()
plt.show()
```
